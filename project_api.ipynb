```python
import requests
import json
```

```python
# base URL for AirBnB data query 

url = "https://public.opendatasoft.com//api/explore/v2.1/catalog/datasets/airbnb-listings/records?refine="

# country filter
country_1st_word = "United"
country_2nd_word = "States"

# city filter
city = "Denver"
```

```python
query_url = f"{url}country%3A%22{country_1st_word}%20{country_2nd_word}%22&refine=city%3A%22{city}%22"
```

```python
response_data = requests.get(query_url)
print(response_data)
```
